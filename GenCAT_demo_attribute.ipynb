```python
import gencat
import numpy as np
from scipy.stats import wasserstein_distance
import matplotlib.pyplot as plt
```

```python
def att_plot(X,Label,tag):
    plt.rcParams["font.size"] = 21
    fig = plt.figure(figsize=(7, 7))
    ax = fig.add_subplot(1,1,1)
    colors = ['red','blue','green','purple','gold','brown','c','m','k','plum','yellow','pink','maroon','teal','tomato']
    markers = ['.',',','v','^']

    partition = []
    for i in range(k):
        partition.append([])
    for i in range(len(Label)):
        partition[Label[i]].append(i)

    count = 1
    for i in partition:
        tmp_ver = []
        tmp_hor = []
        for j in i:
            tmp_ver.append(X[j,0])
            tmp_hor.append(X[j,1])
#         ax.scatter(tmp_ver,tmp_hor, c=colors[count],label=count, s=6, marker=markers[count])
        ax.scatter(tmp_ver,tmp_hor, c=colors[count-1],label=count, s=1.5)
        count+=1
        if count == 5: # how many classes do you want to plot?
            break

    plt.xlabel("attribute1", size=32)
    plt.ylabel("attribute2", size=32)
    plt.legend(bbox_to_anchor=(0.45, 1.0), loc='lower center', borderaxespad=1., ncol=4 , markerscale=10., scatterpoints=1, fontsize=18,title='class').get_title().set_fontsize(30)
    plt.tight_layout()
    plt.show()
    
    
def hist_plot(X,tag):
    plt.rcParams["font.size"] = 16
    fig = plt.figure()
    # Axesを追加
    ax = fig.add_subplot(111)
    count, bins, ignored = ax.hist(X[:,0], 100, density=True)
    ax.set_xlabel("attribute value", size = 28)
    ax.set_ylabel("Frequency", size = 28)
    plt.tight_layout()
    # # pdfファイルの初期化
    # pp = PdfPages('attribute_hist.pdf')
    # # figureをセーブする
    # pp.savefig()
    # # pdfファイルをクローズする。
    # pp.close()
#     fig.savefig("fig_information_systems/hist_plot_"+tag+".png",dpi=200)
    plt.show()
    
def emd(X,omega):
    tmp0 = np.random.normal(np.average(X[:,0]),omega,n)
    tmp1 = np.random.normal(np.average(X[:,1]),omega,n)
    return wasserstein_distance(X[:,0],tmp0) + wasserstein_distance(X[:,1],tmp1)
```

```python
n=5000
m=20000
max_deg = 500
k=4
d=2
omega=0.2
            
M = np.array([
    [0.7, 0.1, 0.1, 0.1],
    [0.1, 0.7, 0.1, 0.1],
    [0.1, 0.1, 0.7, 0.1],
    [0.1, 0.1, 0.1, 0.7],
])
            
D = np.array([
    [0.2, 0.1, 0.1, 0.1],
    [0.1, 0.2, 0.1, 0.1],
    [0.1, 0.1, 0.25, 0.1],
    [0.1, 0.1, 0.1, 0.25],
])
```

# Varying attribute-class correlations

```python
H = np.array([
    [0.5, 0.0, 0.0, 0.5],
    [0.0, 0.5, 0.0, 0.5],
])
S,X,X_not,Label = gencat.gencat_only_att(n,m,k,d,max_deg,M,D,H,omega=omega)
att_plot(X,Label,tag="05")
hist_plot(X,tag="05")


print("EMD for applying distribution :    ",emd(X,omega))
print("EMD for NOT applying distribution :",emd(X_not,omega))
```
