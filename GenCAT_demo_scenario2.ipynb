# Demo: Reproduction

```python
import gencat
import numpy as np

from func import adj_plot,load_data
```

## load dataset

```python
file_names = ['./datasets/cora_ml.npz']
file_name = file_names[0]

S_ori,C,n,m,k = load_data(file_name)
```

```python
d = 2 # number of attributes

# manual input : attribute-class correlation
H = np.array([
    [0.5, 0.0, 0.0, 0.5, 0.0, 0.0, 0.0],
    [0.0, 0.5, 0.0, 0.5, 0.0, 0.0, 0.0],
])
S,X,Label = gencat.gencat_reproduction(S_ori,C,H,d)
```

```python
adj_plot(S,Label)
```

# changing size

```python
theta = np.zeros(len(C))
nnz = S_ori.nonzero()
for i in range(len(nnz[0])):
    if nnz[0][i] < nnz[1][i]:
        theta[nnz[0][i]] += 1
        theta[nnz[1][i]] += 1
maxdeg = max(theta)

n=6000
m=20000
d=2
H = np.array([
    [0.5, 0.0, 0.0, 0.5, 0.0, 0.0, 0.0],
    [0.0, 0.5, 0.0, 0.5, 0.0, 0.0, 0.0],
])

S,X,Label = gencat.gencat_reproduction(S_ori,C,H,d,n=n,m=m,max_deg=int(maxdeg*2))
adj_plot(S,Label)
```
